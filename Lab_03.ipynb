```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Prepare a sample corpus
corpus = [
    "The quick brown fox jumps over the lazy dog.",
    "Never jump over the lazy dog quickly.",
    "A fox is quick and a dog is lazy.",
    "The dog barked at the quick fox."
]
```

```python
# Step 3: Convert text to Bag-of-Words representation
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(corpus)
```

```python
# Step 4: Visualize the Term-Document Matrix
# Convert the sparse matrix to a dense format and create a DataFrame
df = pd.DataFrame(X.toarray(), columns=vectorizer.get_feature_names_out(), index=[f"Doc{i+1}" for i in range(len(corpus))])

print("Term-Document Matrix:")
print(df)
```

```python
# Step 5: Analyze and visualize the term frequencies
term_frequencies = np.sum(X.toarray(), axis=0)
terms = vectorizer.get_feature_names_out()

# Plot term frequencies
plt.figure(figsize=(10, 6))
sns.barplot(x=terms, y=term_frequencies)
plt.xticks(rotation=45)
plt.title("Term Frequencies in the Corpus")
plt.xlabel("Terms")
plt.ylabel("Frequencies")
plt.show()
```
